```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def read_world_bank_data(file_path):
    """
    Read World Bank data from a CSV file and perform data transformation.

    Parameters:
    - file_path (str): Path to the CSV file containing World Bank data.

    Returns:
    - df_years (DataFrame): Transposed dataframe with years as columns.
    - df_countries (DataFrame): Dataframe with countries as columns.
    """
    # Read the World Bank format dataframe
    df = pd.read_csv(file_path)

    # Transpose the dataframe to have years as columns
    df_years = df.set_index('Country Name').T.reset_index()

    # Clean the transposed dataframe
    df_years.columns.name = None  # Remove the column name
    df_years = df_years.rename(columns={'index': 'Year'})  # Rename the index column to 'Year'

    # Create a dataframe with countries as columns
    df_countries = df.set_index('Country Name')

    return df_years, df_countries

def plot_forest_area_correlation_heatmap(dataframe):
    """
    Plot a correlation heatmap for forest area (% of land area) for selected countries.

    Parameters:
    - dataframe (pd.DataFrame): The transposed DataFrame containing forest area data.

    Returns:
    - None
    """

    # Specify the countries to select
    selected_countries = ['Pakistan', 'Bangladesh', 'Canada', 'France', 'China',
                           'United Kingdom', 'India', 'United States', 'Brazil', 'Sweden']

    # Extract data for the specified years and selected countries
    years_to_plot = [str(year) for year in range(1990, 2021, 5)]
    data_to_plot = dataframe.loc[selected_countries, years_to_plot]

    # Calculate the correlation matrix
    correlation_matrix = data_to_plot.T.corr()

    # Plotting the heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=.5)
    plt.title('Correlation Heatmap for Forest Area (% of Land Area)')
    plt.show()

# Assuming you have the transposed DataFrames: countries_df_forest_area
# If not, replace it with the appropriate DataFrame
# Call the function with your DataFrame
# plot_forest_area_correlation_heatmap(countries_df_forest_area)

def plot_arable_land(data, selected_countries, years_to_plot):
    """
    Plot a stacked bar graph for arable land in selected countries over the specified years.

    Parameters:
    - data (pd.DataFrame): The transposed DataFrame containing arable land data.
    - selected_countries (list): List of countries to be selected for plotting.
    - years_to_plot (list): List of years to be plotted.

    Returns:
    - None
    """

    # Extract data for the specified years and selected countries
    data_to_plot = data.loc[selected_countries, years_to_plot]

    # Plotting the bar graph
    plt.figure(figsize=(12, 8))
    data_to_plot.T.plot(kind='bar', stacked=True)
    plt.title('Arable Land in Selected Countries (1990 to 2020)')
    plt.xlabel('Year')
    plt.ylabel('Arable Land (% of Land Area)')
    plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()
def plot_co2_emissions(data, selected_countries, years_to_plot):
    # Extract data for the specified years and selected countries
    data_to_plot = data.loc[selected_countries, years_to_plot]

    # Plotting the bar graph
    plt.figure(figsize=(12, 8))
    data_to_plot.T.plot(kind='bar', stacked=True)
    plt.title('CO2 Emissions by Selected Countries (1990 to 2020)')
    plt.xlabel('Year')
    plt.ylabel('CO2 Emissions (metric tons per capita)')
    plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()

# Assuming you have the transposed DataFrames: countries_df_co2_emission
# If not, replace it with the appropriate DataFrame
def plot_data(selected_countries, data, data_label, ylabel):
    # Loop through the selected countries
    for country in selected_countries:
        # Select data for the current country from the dataframes
        country_data = data.loc[country, '1990':'2020']

        # Plotting the data
        plt.figure(figsize=(12, 6))
        plt.plot(country_data, marker='o', label=f'{data_label} - {country}')
        plt.title(f'{data_label} in {country} (1990-2020)')
        plt.xlabel('Year')
        plt.ylabel(ylabel)
        plt.xticks(selected_years)
        plt.legend()
        plt.show()
def plot_power_consumption(data, selected_countries, years_to_plot):
    """
    Plot a stacked bar graph for power consumption by selected countries over specified years.

    Parameters:
    - data (pd.DataFrame): The transposed DataFrame containing power consumption data.
    - selected_countries (list): List of countries to be selected for plotting.
    - years_to_plot (list): List of years to be plotted.

    Returns:
    - None
    """

    # Extract data for the specified years and selected countries
    data_to_plot = data.loc[selected_countries, years_to_plot]

    # Plotting the bar graph
    plt.figure(figsize=(12, 8))
    data_to_plot.T.plot(kind='bar', stacked=True)
    plt.title('Power Consumption by Countries (1990 to 2010)')
    plt.xlabel('Year')
    plt.ylabel('Power Consumption')
    plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()

# Assuming you have the transposed DataFrames: countries_df_power_consumption
# If not, replace it with the appropriate DataFrame
def plot_power_consumption(data, selected_countries, years_to_plot):
    """
    Plot a stacked bar graph for power consumption by selected countries over specified years.

    Parameters:
    - data (pd.DataFrame): The transposed DataFrame containing power consumption data.
    - selected_countries (list): List of countries to be selected for plotting.
    - years_to_plot (list): List of years to be plotted.

    Returns:
    - None
    """

    # Extract data for the specified years and selected countries
    data_to_plot = data.loc[selected_countries, years_to_plot]

    # Plotting the bar graph
    plt.figure(figsize=(12, 8))
    data_to_plot.T.plot(kind='bar', stacked=True)
    plt.title('Power Consumption by Countries (1990 to 2010)')
    plt.xlabel('Year')
    plt.ylabel('Power Consumption')
    plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()

# Assuming you have the transposed DataFrames: countries_df_power_consumption
# If not, replace it with the appropriate DataFrame
# Specify the countries to select


```
